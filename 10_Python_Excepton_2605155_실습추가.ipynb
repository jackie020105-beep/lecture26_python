```python
# 양의 정수를 입력할 때까지 입력을 계속 받기
while True:
    try:
        number = int(input('>> 양의 정수 입력 : '))
        if number <= 0:
            # ValueError 불러오기
            raise ValueError
    except ValueError:
        print('양의 정수를 입력하시오')
    else:
        sum = 0
        for i in range(1, number+1):
            sum += i
        print(f'>> 1부터 {number}까지의 합 : {sum}')
        break
```

```python
# 3. 함수 divide(x,y) 구현에서 연산 x/y은 0 으로 나누는 것에 대한 예외 처리를 수행하도록 한다. 다음 두 문장의 함수 호출에서 다음과 같은 결과가 나오도록 프로그램을 작성하시오.
# - 구문은 try except else 블록 구현
# - 다음 두 호출에 대한 결과
### divide(3.2, 2)
### divide(5.4, 0)

###(결과) 결과: 1.6
###(결과) 0으로는 나눌 수 없습니다


# 정답
def divide(x,y):
    try: 
        if y == 0:
            raise ZeroDivisionError
    except ZeroDivisionError:
        return(print('0으로는 나눌 수 없습니다'))
    else: 
        return(print(f'(결과): {x/y}'))

divide(3.2, 2)
divide(5.4, 0)
```

```python
# 4. 다음 딕셔너리를 사용해 표준 입력에서 받은 정수에 해당하는 영어 문자열을 출력하고, 잘못된 입력에 대해 발생하는 예외를 처리하는 프로그램을 작성하시오.
### uyear = {1:'freshman', 2:'sophmore', 3:'junior', 4:'senior'}
# - 구문은 try except else 블록 구현
# - 1에서 4 사이의 입력이면 정상적으로 실행돼 입력한 대학 학년의 영어 이름이 출력
### 대학교 몇 학년이지요? 1
### 1학년: freshman
### 예외 처리가 잘되는군요!
# - 1에서 4 사이를 벗어나는 정수를 입력하면 블록 try에서 raise로 예외 발생시켜 다음과 같이 출력
### 대학교 몇 학년이지요? 5
### 예외 발생 이름: <class 'Exception'>
### 예외 발생 이유: 1~4 정수를 입력하시오
### 예외 처리가 잘되는군요!


# 정답
uyear = {1:'freshman', 2:'sophmore', 3:'junior', 4:'senior'}
try:
    npt = int(input('대학교 몇 학년이지요? '))
    if not (1 <= npt <= 4):
        raise Exception

except Exception:
    print(f'대학교 몇 학년이지요? {npt}')
    print("예외 발생 이름: <class 'Exception'>")
    print('1~4 정수를 입력하시오')
    print('예외 처리가 잘되는군요!')

else:
    print(f'대학교 몇 학년이지요? {npt}')
    print(f'{npt}학년: {uyear[npt]}')
    print('예외 처리가 잘되는군요!')
```
