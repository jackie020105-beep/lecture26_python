### 파이썬 클래스

```python
# 원에 대한 넓이 구하기
class Circle:
    pi = 3.14
    def __init__(self, radius):
        self.radius = radius
    def area(self):
        return self.radius ** 2 * Circle.pi

c = Circle(2.5)
print(c.area())
```

```python
# 사각형에 대한 넓이 구하기
class Rectangle:    
    def __init__(self, width, height):
        self.width = width
        self.height = height
    def area(self):
        return self.width * self.height

rec = Rectangle(2.1, 3.4)
print(rec.area())
```

```python
# 은행계좌 만들기 1
class Account:
    def __init__(self):
        self.balance = 0
    def deposit(self, amount):
        self.balance += amount
    def withdraw(self, amount):
        self.balance -= amount


a2 = Account()
a2.deposit(50)
print(a2.balance)
```

```python
# 은행계좌 만들기 2
# __balance는 데이터 숨기기(은닉)
class Account:
    def __init__(self):
        self.__balance = 0
    def deposit(self, amount):
        self.__balance += amount
    def withdraw(self, amount):
        self.__balance -= amount
    def get_balance(self):
        return self.__balance

a2 = Account()
a2.deposit(50)
print(a2.get_balance())
```

```python
class Person:
    def __init__(self, name, age):
        self.name = name
        self.age = age
    # 메모리의 주소가 아닌 문자열로 나타내고 싶을때
    def __str__(self):
        return f'{self.name}({self.age}살)'
    # 같은 사람인지 확인
    def __eq__(self, other):
        if self.name == other.name and self.age == other.age:
            return True
        return False


p1 = Person('홍길동', 15)
print(p1)

p2 = Person('홍길동', 15)
print(p2)

if p1 == p2:
    print('같은 사람입니다')
else:
    print('다른 사람입니다')
```

```python
# 다음을 만족하는 클래스 Student를 작성해 다음과 같은 결과가 나오도록 프로그램을 작성하시오

# - 구초기화 메소드의 인자는 이름, 학과, 중간점수, 기말점수를 저장하는 name, dept, mid, final
# - 문자열화 메소드 __str__ 구현
# - 메소드 grade()는 두 점수의 평균이 90이상이면 A, 80이상이면 B, 70이상이면 C, 60이상이면 D, 그 이하이면 F
# - 클래스 Student 객체 하나를 만들어 문자열화 메소드와 grade() 메소드 호출

###(결과) 학과: 기계학과 / 이름: 김경철 / 중간: 89 / 기말: 90
###(결과) 학점: B


# 정답
class Student:
    def __init__(self, name, dept, mid, final):
        self.name = name
        self.dept = dept
        self.mid = int(mid)
        self.final = int(final)

    def grade(self):
        total = (self.mid + self.final) / 2
        if total >= 90:
            return 'A'
        elif total >= 80:
            return 'B'
        elif total >= 70:
            return 'C'
        elif total >= 60:
            return 'D'
        else:
            return 'F'
        
    def __str__(self):
        return f'학과: {self.dept}, 이름: {self.name}, 중간: {self.mid}, 기말: {self.final}'

student = Student('김경철','기계학과', 89, 90)
print(student)
print(f'학점: {student.grade()}')
```

```python
# 클래스 Car에서 다음 조건의 매직 메소드를 만들어 다음과 같은 결과가 나오도록 프로그램읆 작성하시오

# - 초기화 메소드 __init__
# - 문자열화 메소드 __str__
# - 동등 비교 메소드 __eq__
# - 다음을 구현한 결과
### my car = Car('현대', 2020, '검정')
### yourcar = Car('기아', 2021, '백색')
### print(mycar)
### print(yourcar)
### print(mycar == yourcar)

###(결과) 회사: 현대, 년식: 2020, 색상: 검정
###(결과) 회사: 기아, 년식: 2021, 색상: 백색
###(결과) False


# 정답
class Car:
    def __init__(self, brand, year, color):
        self.brand = brand
        self.year = year
        self.color = color
    def __str__(self):
        return f'회사: {self.brand}, 년식: {self.year}, 색상: {self.color}'
    def __eq__(self, other):
        if self.brand == other.brand:
            if self.year == other.year:
                if self.color == other.color:
                    return True
        else:
            return False
    
mycar = Car('현대', 2020, '검정')
yourcar = Car('기아', 2021, '백색')

print(mycar)
print(yourcar)
print(mycar == yourcar)
```
