# 리스트에 항목 추가하기 (append,insert,extend)

```python
# append
test = []
a=10
b=11
c=12
d=13
test.append(a)
test.append(b)
test.append(c)
test.append(d)
print(test)

# 𝒊𝒏𝒔𝒆𝒓𝒕
tlst = [1, 2, 3, 4, 5, 6]
tlst.insert(2, 'three')
print(tlst)

# 𝒆𝒙𝒕𝒆𝒏𝒅
lst = ['who', 'are', 'you', '?']
lsst = ['where', 'are', 'we', '?']
lst.extend(lsst)
print(lst)

lst = [1, 2, 3, 4]
lsst = ['1', '2', '3', '4']
lssst = []
lssst = lst + lsst
print(lssst)

lst = ['H', 'E', 'L', 'L', 'O']
lsst = ['1', '2', '3', '4']
print(lst * 2)
print(3 * lsst)
```

# 리스트에서 항목 삭제하기 (pop)

```python
# pop(index)
tlst = [111, 222, 333, 444, 555, 666]
i = tlst.pop(2)
print(i)

# pop()
tlst = [1, 2, 3, 4, 5, 6]
d = tlst.pop()
print(d)

# remove(item)
lst = ['who', 'are', 'you', '?']
lst.remove('who')
print(lst)

# del
lst = [1, 2, 3, 4]
lsst = ['1', '2', '3', '4']
del lst[2]
print(lst)
del lsst[:2]
print(lsst)
del lsst[::2]
print(lsst)

# clear
lst = ['H', 'E', 'L', 'L', 'O']
print(lst, end='')
lst.clear()
print(lst)
```

# 리스트의 항목 재배열하기 (reverse)

```python
testlst = [1, 2, 3, 4, 5, 6, 7, 8]
testlst[:3] = [111, 222, 333]
testlst.reverse()
print(testlst)
```

```python
alst = ['one','two','three']
print(alst)
alst.reverse()
print(alst)
```

```python
alst = ['one','two','three','four']
print('원본데이터', alst)
alst.reverse()
print('역순데이터',alst)

# sort=아스키코드 알파벳 순서
alst.sort()
print('정렬데이터',alst)

blst = sorted(alst)
print('원본데이터',alst)
print('정렬데이터',blst)

clst = sorted(alst, key=len)
print(clst)
```

```python
scores = [95, 45, 21, 74, 55]
print(sorted(scores))
```

```python
scores = [95, 45, 21, 74, 55]
print(sorted(scores, reverse=True))
```

```python
scores = [95, 45, 21, 74, 55]

scores.sort(reverse=True)
print(scores)
```

# 리스트 탐색하기 (in, index,count,len,max,min,sum)

```python
# index
lst = [11, 22, 33, 44, 55, 66, 77, 88]
tar=66
print(tar in lst, end=',')
print(lst.index(tar))
tar=99
print(tar in lst)

# count
lst = [11, 88, 33, 44, 88, 66, 77, 88]
tar=88
print(lst.count(tar))

# len
lst = [11, 88, 33, 44, 88, 66, 77, 88]
a = len(lst)-2
print(a)

# max
scores = [95, 45, 21, 74, 55]
print(max(scores))

# min
scores = [95, 45, 21, 74, 55]
print(min(scores))

# sum
scores = [1, 2, 3, 4, 5]
print(sum(scores))

```

# 리스트 내포

```python
# 1~N까지의 수로 이루어진 리스트를 생성

n = int(input())
numlst=[]
for i in range(1,n+1):
    numlst.append(i)
print(numlst)
```

```python
# expression

n = int(input())
numlst=[i for i in range(1,n+1)]
print(numlst)
numlst=[i * 2 for i in range(1,n+1)]
print(numlst)
```

```python
# if문 사용

n = int(input())
numlst=[i for i in range(1,n+1) if i % 2 == 0 ]
print(numlst)
```
