# 튜플(tuple) 만들기

```python
s = (1, 2, 3)
print(s)
s = (1, 2, 3, ('a', 'b', 'c'))
print(s)
s = (1, 2, 3, ([1, 2], 'c'))
print(s)
```

```python
# 빈 튜플의 생성 : ( ), tuple()
sq = ()
print(sq)
# 항목이 1개인 튜플의 생성
sw = (1,)
print(sw)
sr = 1,
print(sr)
# 괄호 없는 튜플
se = 1, 2, 3, 4
print(se)

x, y, z, a = se
print(x, y, z, a)
```

# 튜플 (tuple) 다루기

```python
# 튜플 항목에 접근
```

```python
 # Indexing
num = 1234, 2345, 3456, 4567
print(num[0])
print(num[3])
a = 'a','p','p','l','e'
print(a[2])
print(a[0])
print(a[1])
```

```python
# slicing
se = 1, 2, 3, 4
print(se[1:2])
print(se[1:3])
```

```python
# for ~ in 문
se = 1, 2, 3, 4
for i in se:
    print(i)
se = '정원재'
for i in se:
    print(i)
se = '정''원''재'
for i in se:
    print(i, end='')
```

```python
# 튜플 탐색
```

```python
# in 연산자
se = (1, 2, 3, 2, 4)
print(2 in se)
print(5 in se)
```

```python
# index()
se = (1, 2, 3, 2, 4)
print(se.index(2))
print(se.index(3))
```

```python
# count()
se = (1, 2, 3, 2, 2, 4)
print(se.count(2))
```

```python
# 튜플 관련 정보
```

```python
# index()
se = 1, 2, 3, 4
print(len(se))
```

```python
# max()
se = 5, 8, 3, 4
print(max(se))
```

```python
# min()
se = 5, 8, 3, 4
print(min(se))
```

```python
# sum()
se = 5, 8, 3, 4
print(sum(se))
```

```python
# 새로운 튜플 생성
```

```python
# sorted
se = (3, 1, 4, 2)
print(sorted(se))
```

```python
# + 연산
se = (1, 2)
sr = (3, 4)
print(se + sr)
```

```python
# * 연산
se = (1, 2)
print(se * 3)
```

```python
# 튜플 전체 삭제
```

```python
# del 
se = (3, 1, 4, 2)
print(se)
del se
# NameError
print(se)
```

# 튜플의 활용

```python
# 두 변수의 값 바꾸기 (swap)

a, b, c = 10, 100, 1000
a, b, c = c, a, b
print(a, b, c)
```

```python
# 함수에서 여러 개의 값 반환
a=10
b=3
def add(a, b):
    return a+b, a*b

re = add(a, b)
print(re)

a=10
b=3
c=20
def add(a, b, c):
    return a+b*c, a*b+c

re = add(a, b, c)
print(re)
```
