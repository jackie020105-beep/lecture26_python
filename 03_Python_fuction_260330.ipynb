```python
# num에 해당하는 구구단 출력

def gugudan(num):
    print(f' [구구단 {num}단]')
    for i in range(1,10):
        print(f'{num} X {i} = {num*i}')
```

```python
number = int(input("단 입력"))
gugudan(number)
```

```python
# return 값 확인

def add_and_mul(a, b):
    return a+b, a*b
    
result = add_and_mul(3,4)
print(result)

result1, result2 = add_and_mul(3,4)
print(result1, result2)
```

```python
# 지역 변수, 전역 변수 (오류나는 경우)

sum = 10
def test(num):
    sum += num
    
test(5)
print(sum)
```

```python
# 지역 변수, 전역 변수 (추천)

sum = 10
def test(sum, num):
    sum += num
    return sum
    
sum = test(sum, 5)
print(sum)
```

```python
# 지역 변수, 전역 변수 (비추천: 파라미터를 쓸 수 있는데 global을 쓰면 범용성이 안 좋음)

sum = 10
def test(num):
    global sum
    sum += num
    
test(5)
print(sum)
```
