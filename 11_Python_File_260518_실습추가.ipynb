```python
# 2. 모듈 math와 random을 불러와 random()함수로 발생한 값을 활용해 다음과 같이 원의 면적을 출력하는 프로그램을 작성하시오.
# - 모듈 math의 원주율 pi 사용
# - 모듈 random의 함수 random()*10으로 반지름 사용
# - 반지름은 내장 함수 round()로 소수점 두 자리 사용
# - 함수 getarea(r)로 원의 면적을 반환
### 원의 반지름: 9.46
### 원주율 pi: 3.1416
### 반지름 9.46인 원의 면적은 281.23


# 정답
import math
import random

r = round(random.random() * 10, 2)
def getarea(r):
    return r * r * math.pi

area = getarea(r)

print(f"### 원의 반지름: {r}")
print(f"### 원주율 pi: {math.pi:.4f}")
print(f"### 반지름 {r}인 원의 면적은 {area:.2f}")
```

```python
# 3. 모듈 math의 factorial() 함수와 모듈 statistics의 median(), mean(), variance(), stdev() 함수를 사용해 다음과 같이 출력하는 프로그램을 작성하시오
# - 모듈로 다시 활용될 수 있도록 다음 조건 블록에 코딩
### if __name__ == '__main__':
# - 다음 리스트 자료의 중앙 값, 평균, 분산, 표준편차를 출력
### st = [80, 99, 77, 65, 92, 74, 82]

###(결과) 1! = 1
###(결과) 6! = 720
###(결과) 11! = 39916800
###(결과) 16! = 20922789888000

###(결과) [80, 99, 77, 65, 92, 82]
###(결과) 중앙 값: 80.00
###(결과) 평균: 81.29
###(결과) 분산: 127.90
###(결과) 표준편차: 11.31


# 정답
import math
import statistics

if __name__ == '__main__':
    for i in range(1, 17, 5):
        print(f'{i}! = {math.factorial(i)}')

    st = [80, 99, 77, 65, 92, 74, 82]
    print(f'{st}')
    print(f'중앙 값: {statistics.median(st)}')
    print(f'평균: {statistics.mean(st)}')
    print(f'분산: {statistics.variance(st)}')
    print(f'표준편차: {statistics.stdev(st)}')
```

### 이미지 파일 복사하기 함수 copy_image_file(image_file)

```python
# - 이미지 파일 복사하기 함수 copy_image_file(image_file) : jpg, png, gif 파일
### 복사 대상 함수 이름은 image_file의 파일명에 ‘_copy’를 추가하고 확장명은 그대로 유지

import os

def copy(image_file):
    name, file = os.path.splitext(image_file)
    copy_file = name + '_copy' + file
    
    with open(image_file, 'rb') as f:
        content = f.read()
        
    with open(copy_file, 'wb') as F:
        F.write(content)
        
    print(copy_file)
```
