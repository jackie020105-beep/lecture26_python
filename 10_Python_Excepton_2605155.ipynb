```python
try:
    num_list = [2, 6, 34, 65, 11]
    index = int(input('>> 인덱스 입력 : '))
    print(num_list[index])
except ValueError:
    print(f'0부터 {len(num_list)-1}까지의 정수를 입력하시오')
except IndexError:
    print(f'0부터 {len(num_list)-1}까지의 정수를 입력하시오')
# 모든 예외 처리(가장 상위 옵션)
except Exception as e:
    print(f'예외가 발생했습니다 : {e} -> {type(e)}')
```

```python
try:
    num_list = [2, 6, 34, 65, 11]
    index = int(input('>> 인덱스 입력 : '))
    print(num_list[index])
# 예외 처리 묶어서 사용 가능
except (ValueError, IndexError):
    print(f'0부터 {len(num_list)-1}까지의 정수를 입력하시오')
# 모든 예외 처리(가장 상위 옵션)
except Exception as e:
    print(f'예외가 발생했습니다 : {e} -> {type(e)}')
```

```python
# 양의 정수를 입력할 때까지 입력을 계속 받기
while True:
    try:
        number = int(input('>> 양의 정수 입력 : '))
        if number <= 0:
            # ValueError 불러오기
            raise ValueError
    except ValueError:
        print('양의 정수를 입력하시오')
    else:
        sum = 0
        for i in range(1, number+1):
            sum += i
        print(f'>> 1부터 {number}까지의 합 : {sum}')
        break
```

```python
def divide(x,y):
    try: 
        if y == 0:
            raise ZeroDivisionError
    except ZeroDivisionError:
        print('0으로는 나눌 수 없습니다')
    else: 
        print(f'(결과): {x/y}')



divide(3.2, 2)
divide(5.4, 0)
```

```python
# 정답
uyear = {1:'freshman', 2:'sophmore', 3:'junior', 4:'senior'}
try:
    npt = int(input('대학교 몇 학년이지요? '))
    if not (1 <= npt <= 4):
        raise Exception

except Exception:
    print(f'대학교 몇 학년이지요? {npt}')
    print("예외 발생 이름: <class 'Exception'>")
    print('1~4 정수를 입력하시오')
    print('예외 처리가 잘되는군요!')

else:
    print(f'대학교 몇 학년이지요? {npt}')
    print(f'{npt}학년: {uyear[npt]}')
    print('예외 처리가 잘되는군요!')
```
